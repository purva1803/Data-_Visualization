# Data Cleaning

---



**Dropping columns Indicator Name and Indicator Code**

```python
population.drop(['Indicator Name','Indicator Code'],axis = 1, inplace = True)
```

```python
population.isna().sum()
```

**Replacing mean values of the columns with the NaN values**

```python
years = list(map(str, range(1960, 2017)))
population[years] = population[years].round(decimals=1).astype('float64')
population[years] = population[years].apply(lambda col: col.fillna(col.mean()), axis=0)
population.head()
```

```python
population.isna().sum()
```

```python
population.info()
```

**Melting the dataset**

```python
df = pd.melt(population,
        id_vars='Country Code',
        value_vars=years,
        var_name='Year',
        value_name='Population')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df_merged = pd.merge(country,df,how='left',on='Country Code')
df_merged.head()
```
